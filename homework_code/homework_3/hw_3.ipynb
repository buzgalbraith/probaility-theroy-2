```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
data=pd.read_csv('/mnt/c/Users/buzga/Desktop/School/grad_school/spring_2023/prob_stats_2/homework/homework_3/cardio.csv' ,sep=";")
data=data["ap_hi"]
temp=np.random.choice(data, int(len(data)*.01))
bias= lambda x: np.mean(x)-np.mean(data)
def find_bias_variance(start, step):
    out=[[],[]]
    for i in np.arange(start, 1, step):
        x=np.random.choice(data, int(len(data)*i))
        out[0].append(bias(x))
        out[1].append(np.var(x))
    plt.plot(np.arange(start, 1, step), out[0])
    plt.xlabel("percentage  sampled")
    plt.ylabel("bias")
    plt.title("bias of sample for difrent proprtions of the sample")
    plt.show()
    plt.plot(np.arange(start, 1, step), out[1])
    plt.xlabel("percentage  sampled")
    plt.ylabel("variance")
    plt.title("variance of for difrent proprtions of the sample")
    plt.show()
    return out
bias,variance=find_bias_variance(.01,.01)
```
